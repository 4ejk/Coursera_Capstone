## Data Scrapping and Forming a Dataframe

Firstly, we shall load the data . This can be done by extracting the data from the website: *[https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M](https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M)*.           
Now the data available at the website is not in a form that is readily available so as to convert it into a pandas dataframe easily, so, we must scrape the data from the website.     
      
Web scraping or data scraping is used for extracting data from websites. It extracts underlying HTML code and, with it, data stored in a database. The scraper can then replicate entire website content elsewhere.

```python
!pip install requests beautifulsoup4 
```

```python
#Importing necessary libaries for web scrapping

import requests
from bs4 import BeautifulSoup
```

```python
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
# Make a GET request to fetch the raw HTML content
page = requests.get(url).text

# Parse the html content
soup = BeautifulSoup(page, "lxml")
print(soup.prettify()) # print the parsed data of html
```

```python
postable = soup.find("table", attrs={"class": "wikitable sortable"})
postable
```

```python
#for getting the table 
postable_data = postable .tbody.find_all("tr")  
postable_data
```

```python
# Get all the headings of Lists for column names from table
columnam = []
for td in postable_data[0].find_all("th"):
    # remove any newlines and extra spaces from left and right
    columnam.append(td.text.replace('\n', ' ').strip())

print(columnam)
```

```python
tablen = len(postable_data) #length of table
tablen
```

```python
# Geting all rows from table

tabdat=[]
for i in range(1,tablen):
    row=[]
    for td in postable_data[i].find_all("td"):
        # remove any newlines and extra spaces from left and right
        row.append(td.text.replace('\n', ' ').strip())
    tabdat.append(row)      
print(tabdat)
```

```python
print(len(tabdat)) #checking table length 
```

```python
# finally converting data list into dataframe
import pandas as pd
df = pd.DataFrame(data=tabdat,columns=columnam)
df.head()
```

```python
df.shape #checking shape of dataframe
```

Since the credentials are accepted we shall start exploring all neighbourhoods.    
First lets create a function to get the venues in all the neighbourhoods. 

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighbourhood', 
                  'Neighbourhood Latitude', 
                  'Neighbourhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

Now we shall create a dataframe containing all the venues in the different neighbourhoods and put the above function in action.

```python
toronto_venues = getNearbyVenues(names=newtp_df['Neighbourhood'],
                                   latitudes=newtp_df['Latitude'],
                                   longitudes=newtp_df['Longitude']
                                  )
```

```python
print(toronto_venues.shape) # size of new dataframe
toronto_venues.head(7)  
```

```python
toronto_venues.groupby('Neighbourhood').count() # checking returned venues
```

Now lets analyze each neighbourhood

```python
# one hot encoding
tor_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
tor_onehot['Neighbourhood'] = toronto_venues['Neighbourhood'] 

# move neighborhood column to the first column
fixed_columns = [tor_onehot.columns[-1]] + list(tor_onehot.columns[:-1])
tor_onehot = tor_onehot[fixed_columns]

tor_onehot.head()
```

```python
# grouping rows by neighborhood and by taking the mean of the frequency of occurrence of each category

tor_group= tor_onehot.groupby('Neighbourhood').mean().reset_index()
tor_group.head(10)
```

```python
tor_group.shape #checking size of dataframe
```

```python
# a function to sort venues in desending order
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
# A function to create a dataframe with top 5 venues for each neighbourhood

num_top_venues = 5

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Neighbourhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
neighb_venues = pd.DataFrame(columns=columns)
neighb_venues['Neighbourhood'] = tor_group['Neighbourhood']

for ind in np.arange(tor_group.shape[0]):
    neighb_venues.iloc[ind, 1:] = return_most_common_venues(tor_group.iloc[ind, :], num_top_venues)

neighb_venues.head()
```

Now our data is ready and we can start working on clustering them

### Clustering

Use of k-means to cluster the neighborhood into 4 clusters.

```python
# set number of clusters
k = 4

torgroup_clust = tor_group.drop('Neighbourhood', 1)

# run k-means clustering
kmean = KMeans(n_clusters=k, random_state=0).fit(torgroup_clust)

# check cluster labels generated for each row in the dataframe
kmean.labels_[0:10] 
```

Now, lets create a final dataframe that includes cluster and the top 5 venues.

```python
# add clustering labels
neighb_venues.insert(0, 'Cluster Labels', kmean.labels_)

fintoronto = newtp_df

# merge manhattan_grouped with manhattan_data to add latitude/longitude for each neighborhood
fintoronto = fintoronto.join(neighb_venues.set_index('Neighbourhood'), on='Neighbourhood')

fintoronto.head() # check the last columns!
```

Finally, now we shall visualize the clusters

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(fintoronto['Latitude'], fintoronto['Longitude'], fintoronto['Neighbourhood'], fintoronto['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
